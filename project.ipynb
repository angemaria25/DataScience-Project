# Municipio Cerro 

###información extraída de cuba.tesoro

El Cerro es uno de los barrios en el municipio de La Habana. Antiguamente fue barrio extramural de la capital, integrante de su municipio.
Su fundación data del año 1803, cuando empezó a cruzar su actual barriada la Calzada del Cerro. Dos propietarios de predios en aquella localidad, los señores José María Rodríguez y Francisco Betancourt, fabricaron allí sus viviendas, siendo rápidamente su ejemplo seguido por muchos otros.
En 1807 llegó a constituirse en poblado, erigiéndose la primera iglesia de madera, que luego fue reemplazada por la actual, la cual no tardó en ser erigida en la parroquia de El Salvador del Mundo, o como muchos se refieren a ella, la Parroquia del Cerro (a la zona que hoy se le llama El Cerro también se le refirió por San Salvador y por Prensa; este último, también se ha escrito como Prenso, por ser el nombre de un ingenio que había en esta área).
Por unos veinte años, aproximadamente de 1820 a 1840, El Cerro fue el lugar escogido por las familias acomodadas de la capital para pasar el verano.
La Casa de Salud del Centro Asturiano llamada La Covadonga fue una finca de recreo perteneciente a la señora Leonor Herrera.
Los condes de Fernandina, de Santovenia, de Lombillo y de Peñalver levantaron en El Cerro sus quintas. En 1843 existían 5 grandes fincas de recreo, 23 notables por su suntuosidad y 273 particulares de familias acomodadas.
En dicho año su población ascendía a 2,125 habitantes y en 1858 a 2,530 estables. El resto ha sido obra del engrandecimiento de la Ciudad de La Habana, que absorbió al Cerro totalmente.

Se encuentra geográficamente localizado en la porción centro-oeste de la ciudad , tiene una extensión superficial de 10,3 Km². El territorio está conformado por siete Consejos Populares: Latinoamericano, Pilar-Atarés, Cerro, Las Cañas, Palatino, El Canal y Armada.

La industria constituye el peso económico del municipio y dentro de sus producciones fundamentales se cuentan: la producción de cervezas y maltas, jabones y detergentes, tabaquería, productos farmacéuticos, envases de cartón, confección de confituras, y otros renglones.
Son muchas las fábricas y talleres del territorio y entre ellas se destacan: Fábrica de cervezas “Miguel Oramas” (La Polar), Fábrica de vinos “Fortín” (antigua Coca-Cola), dos fábricas de jabones y detergentes, tres laboratorios de productos farmacéuticos, Fábrica de galleticas y chocolates “Gerardo Abreu Fontán” (antigua La Estrella), tres embotelladoras, Empresa cárnica “El Miño”, Fábrica de cigarros “Orlando Nodarse” (H. UPMANN), Ronera Bocoy , Litográfica “La Habana”, Empresa Gráfica “Federico Engels” y la fábrica de pistones “1ro de Mayo” única en su tipo en América Latina.
Es famosa la frase «El Cerro tiene la llave». Se ha fundamentado que el Cerro tiene varias llaves: la del agua, por poseer en su territorio los tres primeros acueductos históricos de La Habana (La Zanja Real y la Presa del Huesillo, el Acueducto Fernando VII y el actual Acueducto de Albear; la de la perfumería, por ser el iniciador de esa industria; y la del deporte, por las instalaciones e instituciones deportivas importantes que posee.

```python
#En esta sección se crea una función para meter los precios de la cerveza en una lista 
products=list(dicc.keys()) #crea una lista con las claves del diccionario dicc

def get(key , info = "precio") : #define una función llamada "get" que toma dos argumentos , "key" e "info" 
    price = []                   #crea una lista llamada "price"
    for element in dicc[key]:    #bucle for que recorre los elementos del valor asociado a la clave key en el diccionario dicc 
        price.append(element[info])  #añade el valor del campo info de cada elemento a la lista price  
    return price                     #devuelve la lista price 

precio_cerveza = get("cerveza")  #llama la función get con el argumento cerveza y guarda el resultado en la variable precio cerveza 
```

```python
#Histogrma para representar los precios de la cerveza y la moda de los precios.
plt.hist(precio_cerveza, bins = 10)     #crea un histograma con los datos de la lista precio cerveza usando 10 intervalos o bins.
plt.axvline(x = moda(precio_cerveza), color = "red") #dibuja una línea vertical roja con el valor de la moda de los precios de la cerveza usando la funcion moda definida anteriormente 
plt.xlabel("Beer Price") #pone una etiqueta al eje x con el texto Beer Price
plt.ylabel("Frequency")  #pone una etiqueta al eje y con el texto Frequency 
plt.title("Histogram of beer prices with mode") #pone titulo al grafico con el texto Histogram of beer prices with mode
plt.show()   #muestra el gráfico
```

```python
 #En esta sección se crea una lista para hallar el % de la cerveza de cada país que se encuentra en las cafeterías .
beer = dicc["cerveza"] #crea una variable llamada “beer” que almaceena el valor de la clave “cerveza” del diccionario “dicc”.
country_of_origin = [] #crea una lista vacía que guardara los paises de procedencia de las cervezas 
for venta in beer:      #bucle for que recorre la lista beer y por cada elemento de la lista agrega el valor de la clave pais de procedencia a la lista country_of_origin
   country_of_origin.append(venta["pais_de_procedencia"]) 

total_paises = len(country_of_origin) #crea una variable llamada “total_paises” que almacena el número de elementos de la lista “country_of_origin”
dicti_of_country = {}                 #crea un diccionario vacío que guardará el conteo de cada país en la lista country_of_origin 

for value in country_of_origin:       #recorre la lista “country_of_origin” con un bucle for  
    if value in dicti_of_country:     #y verifica si ya esta en el diccionario dicti_of_country 
        dicti_of_country[value] += 1  #si esta se le suma uno al valor asociado a esa clave 
    else:                             #sino
        dicti_of_country[value] = 1   #crea una nueva clave con ese país y le asigna el valor uno 


for key in dicti_of_country:           #bucle for que recorre el diccionario “dicti_of_country”
    dicti_of_country[key] = (dicti_of_country[key]/total_paises)*100 #y por cada clave calcula el porcentaje que representa ese país sobre el total de países , divide el valor asociado a esa clave por el valor de total paises y lo multiplica por 100
    

aux = 0                        #crea una variable auxuliar llamada “aux” que almacena el valor cero 
dicti_of_country_aux = {}      #crea otro diccionario vacío que guardará los países con porcentaje mayor o igual a 6 
for item in dicti_of_country:  #bucle for que recorre cada elemento del diccionario “dicti_of_country” y por cada clave verifica
    if dicti_of_country[item] < 6:  #si el valor asociado a esa clave es menor que 6
        aux += dicti_of_country[item] #si es así se suma ese valor a la variable aux 
    else:                            #sino
        dicti_of_country_aux[item] = dicti_of_country[item] #se crea una nueva clave con ese país y su valor en el diccionario dicti_of_country_aux 

dicti_of_country_aux["Other"] = aux   #agrega una nueva clave llamada “Other" al diccionario dicti_of_country_aux con el valor de la variable “aux”
dicti_of_country = dicti_of_country_aux  #reemplaza el valor del diccionario dicti_of_country por el diccionario dicti_of_country_aux 

    
plt.pie(dicti_of_country.values() , labels = dicti_of_country.keys(), autopct ="%1.1f%%") #crea un gráfico circular con los valores  y las claves del diccionario “dicti_of_country” , además usa el argumento autopct para mostrar los porcentajes con decimal en cada porción 
plt.title("Country of origin of beer")  #agrega título al gráfico
plt.show()  #muestra el gráfico con el método show
```

```python
#En esta sección se crea una lista con cada marca de cerveza y su precio promedio , ademas  se halla el precio máximo .
dicti_brand = {}   #crea un diccionario vacío que guardará  las marcas de las cervezas y sus precios 
for value in dicc["cerveza"]:  #crea un bucle for que recorre la lista dicc["cerveza"] y por  elemento (que es otro diccionario) verifica 
    if not (value["marca"] in dicti_brand):  #si la clave "marca" esta en el diccionario  “dicti_brand”, se agrega con su precio como valor
        dicti_brand[value["marca"]] = [value["precio"]] #sino esta crea una nueva clave con esa marca y le asigna una lista con el valor de la clave precio 
    else:         #de lo contrario
        dicti_brand[value["marca"]].append(value["precio"]) #agrega el valor de la clave "precio" a la lista existente 
        

average = {}  #crean un diccionario vacío que guardará las marcas de la cerveza y sus precios promedios 
for marca , precios  in dicti_brand.items(): #crea un bucle for que recorre el diccionario "dicti_brand" y por cada clave (que es una marca) y valor (que es una lista de los precios) usando la función np.mean del módulo numpy
    average[marca] = int(np.mean(precios))  #crea una nueva clave con esa marca y le asigna el valor del promedio convertido a entero 

    
dicti_brand_vs_price_average = {}  #crean un nuevo diccionario vacío que guardará las marcas de cerveza con un precio promedio mayor a 150
for brand , half in average.items():  #crea un bucle for que recorre el diccionario average y por cada clave (que es una marca) y valor (que es un precio promedio) verifica 
    if half > 150 :                   #si el valor es mayor a 150
        dicti_brand_vs_price_average[brand] = half #si es así crea una nueva clave con esa marca y le asigna el valor del precio promedio 
        


color = []  #crea una lista vacía que guardará los colores para el gráfico de barras 
for marca in dicti_brand_vs_price_average: #crea un bucle for que recorre el diccionario "dicti_brand_vs_price_average" y por cada clave (que es una marca) verifica
    if dicti_brand_vs_price_average[marca] == max(dicti_brand_vs_price_average.values()): #si el valor asociado a esa clave es el máximo de los valores del diccionario 
        color.append("#CC99FF") #agrega ese color 
    else:     #sino
        color.append("#6699FF") #agrega este 
dicti_brand_vs_price_average_ord = dict(sorted(dicti_brand_vs_price_average.items(), key=operator.itemgetter(1))) #crea un nuevo diccionario que ordena el diccionario "dicti_brand_vs_price_average" por sus valores usando la función sorted y el módulo operator  
plt.barh(list(dicti_brand_vs_price_average_ord.keys()), dicti_brand_vs_price_average_ord.values(), color = color) #crea un gráfico de barras horizontales usando las claves y los valores del diccionario ordenado como los ejes del gráfico y los colores de la lista "color" como los colores de las barras 
plt.ylabel("Beer Brand")  #título etiqueta 
plt.xlabel("Price")       #título etiqueta 
plt.title("Histogram of brand with price") #título gráfico
plt.show() #muestra el gráfico
```

```python
#En esta sección se calcula la cantidad de envases refrescos de los diferentes tipos 
envase_refresh = {}  #se crea un diccionario vacío 
for value1 in dicc["refresco"]:  #se itera sobre los valores del diccionario "dicc" bajo la clave "refresco"
    for value in value1:   #para cada valor del "refresco" se itera sobre cada clave-valor del diccionario interno
        if value == "envase" and value1["envase"] in envase_refresh: #si la clave es "envase" y el valor correspondiente ya existe en el diccionario  "envase_refresh"
            envase_refresh[value1[value]] += 1  # se incrementa el conteo para ese tipo de envase 
        elif value == "envase":          # si el valor correspondiente no existe en el diccionario
            envase_refresh[value1[value]] = 1  #se crea una entrada con un conteo inicial de 1 
print(envase_refresh)  #finalmente se imprime el diccionario "envase_refresh"

```
